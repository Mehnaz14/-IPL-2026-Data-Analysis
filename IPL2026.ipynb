Feature engineering- creating new columns

```python
df['Total_runs']=df['runs_of_bat']+df['extras'] #total run on each ball
df.head()
```

```python
df['is_six']=0
df['is_four']=0
df['is_wicket']=0
df.loc[df['runs_of_bat']==6, 'is_six']=1
df.loc[df['runs_of_bat']==4, 'is_four']=1
df.loc[df['wicket_type'].notnull(), 'is_wicket']=1
df.head(20)
```

```python
df['over']=df['over']+1
df.head()
```

```python
def get_phase(over):
    if over<=6:
        return 'Powerplay(1-6)'
    elif over<=15:
        return 'Middle(7-15)'
    else:
        return 'Death(16-20)'
df['phase_group']=df['over'].apply(get_phase)
df
```

Strike rate for the top batsman

1. formulae : (runs score / ball faced) * 100
2. wide ball != balls faced

```python
# Filter only legal deliveries
legal_balls = df[df['extras'] == 0]

# Groupby on legal balls
df1 = legal_balls.groupby('striker').agg(
    runs=('runs_of_bat', 'sum'),
    balls=('runs_of_bat', 'count')
).reset_index()

# Filter players
df1 = df1[df1['balls'] > 30]

# Strike rate
df1['strike_rate'] = (df1['runs'] / df1['balls']) * 100

# Sort and take top 10
df1 = df1.sort_values('strike_rate', ascending=False).head(10)

# Clean values
df1['strike_rate'] = df1['strike_rate'].round(1)

df1
```

```python
plt.figure(figsize=(20,6))
plt.bar(df1['striker'], df1['strike_rate'], color='red', edgecolor='black')
plt.title('Top Strikers')
plt.xlabel('Players')
plt.ylabel('strike rate')

for x,y in zip(df1['striker'], df1['strike_rate']):
    plt.text(x=x, y=y, s=y, va='bottom')

plt.show()
```

Average Run rate over by over

```python
# Create over number
df['over_number'] = df['over'].astype(int)

# Group by overs
edf = df.groupby('over_number').agg(
    avg_runs=('Total_runs', 'mean')
).reset_index()

# Round values (apply to edf, not df1)
edf['avg_runs'] = edf['avg_runs'].round(2)

edf
```

```python
plt.figure(figsize=(20,6))

plt.plot(edf['over_number'], edf['avg_runs'],
         marker='o', linewidth=2)

plt.title('Average Runs per Over')
plt.xlabel('Over')
plt.ylabel('Average Runs')

plt.xticks(np.arange(1, 21))

# Add labels
for x, y in zip(edf['over_number'], edf['avg_runs']):
    plt.text(x=x, y=y, s=round(y,2), va='bottom', ha='center')

plt.grid()
plt.show()
```
